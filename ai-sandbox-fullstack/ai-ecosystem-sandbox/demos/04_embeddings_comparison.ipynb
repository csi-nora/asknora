# Demo 04 — Embedding Comparison

Compare **Sentence-Transformers** vs **BGE** (and cloud backends when API keys are set).


```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('.').resolve().parent))

from src.providers.embeddings import embed_texts, cosine_similarity

texts = [
    'Vector database for semantic search',
    'Embedding store for RAG retrieval',
    'Recipe for chocolate cake',
]
for backend in ['sentence-transformers', 'bge']:
    res = embed_texts(texts, backend=backend)
    print(backend, res.model, res.dimensions)
    print('  sim(0,1):', round(cosine_similarity(res.vectors[0], res.vectors[1]), 4))
    print('  sim(0,2):', round(cosine_similarity(res.vectors[0], res.vectors[2]), 4))
```

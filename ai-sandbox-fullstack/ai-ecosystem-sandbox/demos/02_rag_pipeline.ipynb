# Demo 02 — RAG Pipeline

Document load → chunk → embed → **Qdrant** → LLM answer.


```python
import sys
from pathlib import Path
ROOT = Path('.').resolve().parent
sys.path.insert(0, str(ROOT))

from src.providers.rag import load_documents, chunk_documents, index_to_qdrant, rag_query

docs = load_documents(ROOT / 'data/sample_docs')
chunks = chunk_documents(docs)
print(f'Chunks: {len(chunks)}')
index_to_qdrant(chunks)
```

```python
q = 'What is Hybrid RAG and which local endpoint does the lab use?'
out = rag_query(q, backend='qdrant')
print('Answer:', out['answer'])
for i, c in enumerate(out['contexts'], 1):
    print(f'[{i}]', c[:200], '...')
```
